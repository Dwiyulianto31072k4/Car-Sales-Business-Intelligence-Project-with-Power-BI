## Deklarasi Nama State

```python
import pandas as pd


data['state'] = data['state'].replace({
    'ca': 'California',
    'tx': 'Texas',
    'pa': 'Pennsylvania',
    'mn': 'Minnesota',
    'az': 'Arizona',
    'wi': 'Wisconsin',
    'tn': 'Tennessee',
    'md': 'Maryland',
    'fl': 'Florida',
    'ne': 'Nebraska',
    'nj': 'New Jersey',
    'nv': 'Nevada',
    'oh': 'Ohio',
    'mi': 'Michigan',
    'ga': 'Georgia',
    'va': 'Virginia',
    'sc': 'South Carolina',
    'nc': 'North Carolina',
    'in': 'Indiana',
    'il': 'Illinois',
    'co': 'Colorado',
    'ut': 'Utah',
    'mo': 'Missouri',
    'ny': 'New York',
    'ma': 'Massachusetts',
    'pr': 'Puerto Rico',
    'or': 'Oregon',
    'la': 'Louisiana',
    'wa': 'Washington',
    'hi': 'Hawaii',
    'qc': 'Quebec',
    'ab': 'Alberta',
    'on': 'Ontario',
    'ok': 'Oklahoma',
    'ms': 'Mississippi',
    'nm': 'New Mexico',
    'al': 'Alabama',
    'ns': 'Nova Scotia'
})

# Menghapus baris dengan nilai NA di kolom 'state'
data = data.dropna(subset=['state'])

# Menampilkan hasil
print(data['state'].unique())
```

## Menghapus State yang Invalid

```python


# Misalkan data Anda disimpan dalam dataframe bernama df
# Buat filter untuk mendeteksi data yang tidak sesuai (VIN panjang 17 karakter)
invalid_states = data['state'].str.match(r'^[0-9a-zA-Z]{17}$')

# Memindahkan VIN ke kolom 'vin' (jika memang ada kolom 'vin')
data.loc[invalid_states, 'vin'] = data.loc[invalid_states, 'state']

# Menghapus atau mengganti nilai di kolom 'state' yang tidak valid
data.loc[invalid_states, 'state'] = pd.NA  # Atau gunakan 'Unknown' sebagai placeholder
```

```python
# Menampilkan hasil
print(data['state'].unique())
```

```python
# Menghapus baris dengan nilai NA di kolom 'state'
data = data.dropna(subset=['state'])
```

```python
# Menampilkan hasil
print(data['state'].unique())
```

```python
# Menambahkan kolom 'Country' untuk membedakan AS dan Kanada
data['Country'] = data['state'].apply(lambda x: 'United States' if x not in ['Quebec', 'Alberta', 'Ontario', 'Nova Scotia'] else 'Canada')

# Menampilkan data baru dengan kolom Country
print(data[['state', 'Country']].head())
```

## Pembersihan Data (Remove Missing Value)

```python
# Menghapus baris yang memiliki missing value di dataframe 'data'
data = data.dropna()

# Mengecek kembali apakah masih ada missing value
missing_values_after_cleaning = data.isnull().sum()
print(missing_values_after_cleaning)
```
